# Visualizing with Matplotlib

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
mpl.style.use(['ggplot'])
```

```python
haiti = df_can.loc['Haiti', years].copy()
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Imigration from Haiti')
plt.ylabel('Number of imigrants')
plt.xlabel('Years')
plt.show()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Imigration from Haiti')
plt.ylabel('Number of imigrants')
plt.xlabel('Years')
plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

```python
China_India = df_can.loc[['China','India'], years].copy()
China_India
```

```python
China_India.plot(kind='line')
```

```python
df_CI = China_India.transpose()
df_CI.head()
```

```python
df_CI.plot(kind='line')
plt.title("Imigrants from China and India")
plt.ylabel("Number of Imigrant")
plt.xlabel('Years')
plt.show()
```

```python
df_can.sort_values(by='Total', ascending = False, axis=0, inplace=True)
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14,8))
plt.title("Imigrants from China and India")
plt.ylabel("Number of Imigrant")
plt.xlabel('Years')
plt.show()
```

# Area Plots

```python
df_can.sort_values(by='Total', ascending = False, axis=0, inplace=True)
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
```

```python
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area', stacked = False, figsize=(20,10))
plt.title("Imigrants from China and India")
plt.ylabel("Number of Imigrant")
plt.xlabel('Years')
plt.show()
```

```python
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area', alpha=0.25, stacked = False, figsize=(20,10))
plt.title("Imigrants from China and India")
plt.ylabel("Number of Imigrant")
plt.xlabel('Years')
plt.show()
```

```python
ax = df_top5.plot(kind='area',  alpha=0.25, figsize=(20,10))

ax.set_title("Imigrants from China and India")
ax.set_ylabel("Number of Imigrant")
ax.set_xlabel('Years')
```

```python
df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()

df_least5.plot(kind='area', alpha=0.25, figsize=(20,10))
plt.title("Imigrants Trend of Least 5 Countries")
plt.ylabel("Number of Imigrant")
plt.xlabel('Years')

plt.show()
```

```python
ax = df_least5.plot(kind='area', alpha=0.25, figsize=(20,10))
ax.set_title("Imigrants Trend of Least 5 Countries")
ax.set_ylabel("Number of Imigrant")
ax.set_xlabel('Years')
```

# bar Chart

```python
df_iceland = df_can.loc['Iceland', years].copy()
```

```python
df_iceland.plot(kind='bar', figsize=(10,6), rot=90)
plt.title('Imigrant to Canada from 1980 to 2013')
plt.ylabel('NUmber of Imigrant')
plt.xlabel('year')

plt.annotate('',
             xy=(32, 70),
             xytext=(28,20),
             xycoords='data',
             arrowprops=dict(arrowstyle='fancy', connectionstyle='arc3', color='blue', lw=2)
)
plt.annotate('2008-2011 Financial Crises', xy=(28, 30), rotation=72.5, va='bottom',ha='left')

plt.show()
```

# Pie Chart

```python
df_continent = df_can.groupby('Continent', axis=0).sum()
df_continent
```

```python
df_continent['Total'].plot(kind='pie',figsize=(5, 6), autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Immigration to Canada by COntinent')
plt.axis('equal')
plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1,0,0,0,0.1,0.1]
df_continent['Total'].plot(kind='pie',
                          figsize=(15,6),
                          autopct='%1.1f%%',
                          startangle=90,
                          shadow=True,
                          labels=None,
                          pctdistance=1.12,
                          colors=colors_list,
                          explode=explode_list)
plt.title('Immigration to Canada by Continent')
plt.axis('equal')
plt.legend(labels=df_continent.index,loc='upper left')
plt.show()
```
